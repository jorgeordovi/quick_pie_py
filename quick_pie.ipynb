```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
elabels = ['A','B','C','D']
sizes = [15,30,45,10]
```

```python
plt.pie(sizes, labels=elabels)
plt.title('Graph in Pie Basic Stuff')
plt.show
```

```python
plt.style.use('default')
plt.pie(sizes, autopct='%1.1f%%', labels=elabels,  )
plt.title('Graph in Pie Basic Stuff')
plt.show
```

```python
ecolors = ['gold','yellowgreen','lightcoral','lightskyblue']
plt.pie(sizes, labels=elabels, colors=ecolors, autopct='%1.1f%%', startangle=140)
plt.title('Prsonalized Pie Chart')
plt.show
```

```python
## Explosion on the first sector

explosion=(0.1,0,0,0,)
plt.pie(sizes, labels=elabels, colors=ecolors, autopct='%1.1f%%', startangle=140, explode=explosion)
plt.title('Explosion Title On Specicfic Sector')
plt.show()
```
